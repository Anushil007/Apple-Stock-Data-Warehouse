We have got the data till 2023-5-12. But, it is better to update the data warehouse once in a while. So, let's update the data warehouse with the latest data every month.

```python
import pandas as pd
import csv
import yfinance as yf
```

Let's collect the data for 30 days and update the merged csv file.

```python
stock = yf.Ticker("AAPL")
stock.info
```

Out of all this, we need date, open, high, low, adjclose as close and volume.

```python
# get date, open, high,low, adj close as close and volume for AAPL stock with time interval of 1 day for the last 1 month
hist = stock.history( period='1mo')
data = hist[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
data
```

```python
data = data.reset_index()
```

```python
data
```

```python
data.info()
```

Chanage the column's name to make it same as the merged csv file.

```python
data.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
```

```python
data.date.head(1)
```

The date has timestamp as well which is not required. So, let's remove it.

```python
data.date = data.date.dt.date
```

```python
data.date.head(1)
```

```python
data.date.info()
```

Date is now an object. Let's convert it to datetime.

```python
data.date = pd.to_datetime(data.date)
```

```python
data.date.head(1)
```

```python
data.info()
```

```python
data.to_csv('latest.csv',index = False)
```

Merge the latest values into the already present merged CSV file

```python
merged = pd.read_csv('merged.csv')
```

```python
merged.info()
```

```python
merged.date = pd.to_datetime(merged.date)
```

```python
merged.info()
```

```python
merged.head()
```

```python
merged = pd.concat([merged,data],ignore_index = True)
```

```python
merged.info()
```

Check for duplicates. There will be duplicates now as I have already scraped the latest data. Next month, there might be a few of no duplicates.

```python
merged.date.duplicated().sum()
```

Let's remove this duplicate values.

```python
# remove duplicates
merged.drop_duplicates(subset='date', inplace=True)
```

```python
merged.info()
```

```python
merged.to_csv('merged.csv',index = False)
```
